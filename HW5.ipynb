```
class Solution(object):
    def judgeCircle(self, moves):
        """
        :type moves: str
        :rtype: bool
        """
        X_count=Y_count=0
        
        for i in range(len(moves)):
            if moves[i] =='U':
                 X_count+=1
            elif moves[i] =='D':
                 X_count-=1
            elif moves[i] =='R':
                Y_count+=1
            elif moves[i] =='L':
                Y_count-=1
                
        return X_count==Y_count==0
```
