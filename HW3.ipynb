```
class Solution(object):
    def sortArrayByParity(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        s=[]
        for i in A:
            #print(i)#4
            if i%2 == 0:
                s.insert(0, i)
            else:
                s.append(i)
              
        return s
```
